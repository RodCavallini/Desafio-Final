```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
```

### Corretoras por Estado

```python
conn= sqlite3.connect("coderhouse.db")
query = "Select uf as Estado,Count(nome_comercial) as Quantidade from corretoras group by uf order by Quantidade DESC"
corretoras_estado = pd.read_sql_query(query,conn)
print(corretoras_estado)
conn.close()
```

```python
x = corretoras_estado["Estado"]
y = corretoras_estado["Quantidade"]
fig, ax = plt.subplots()
bar_container = ax.bar(x,y)
ax.set_title("Corretoras por Estado", fontsize=16)
ax.set_xlabel("Estados", fontsize=12)
ax.set_ylabel("Quantidade de Corretoras", fontsize=12)
ax.tick_params(axis='both', which='major', labelsize=10)
plt.xticks(rotation=45)
plt.show(ax.bar_label(bar_container))
```

### Corretoras pelo Ano Inicio

```python
conn= sqlite3.connect("coderhouse.db")
query = "Select Ano_Inicio_situacao as Ano_Inicio,Count(nome_comercial) as Quantidade from corretoras group by Ano_Inicio order by Ano_Inicio"
corretoras_estado = pd.read_sql_query(query,conn)
print(corretoras_estado)
conn.close()
```

```python
x = corretoras_estado["Ano_Inicio"]
y = corretoras_estado["Quantidade"]
fig, ax = plt.subplots()
bar_container = ax.bar(x,y,width=0.4)
ax.set_title("Corretoras por Ano Inicio", fontsize=16)
ax.tick_params(axis="both", which='major', labelsize=10)
plt.xticks(rotation=80)
plt.show(ax.bar_label(bar_container))
plt.show()
```
